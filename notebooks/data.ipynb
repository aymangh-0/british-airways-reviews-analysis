الكلمات المتكررة في النصوص

```python
stop_words = {
    "the", "and", "to", "of", "a", "in", "was", "were", "is", "it", "for",
    "on", "with", "that", "this", "as", "at", "we", "i", "my", "our", "they",
    "you", "but", "had", "have", "has", "be", "been", "are", "not", "from",
    "by", "or", "an", "so", "if", "all", "very", "there", "their", "them",
    "he", "she", "his", "her", "me", "us", "will", "would", "could", "should",
    "can", "just", "do", "did", "does", "no", "yes"
}

custom_stop_words = {
    "flight", "flights", "british", "airways", "airway", "ba", "london",
    "which", "when", "one", "only", "get", "also", "even",
    "back", "way", "much", "many", "really",
    "out", "after", "first", "more", "than", "other", "heathrow",
    "class", "business", "economy", "cabin", "return", "check"
}

stop_words = stop_words.union(custom_stop_words)

all_words = []

for review in clean_df["Clean_Review"]:
    review = re.sub(r"[^a-zA-Z\s]", " ", review)
    words = review.lower().split()

    for word in words:
        if word not in stop_words and len(word) > 2:
            all_words.append(word)

word_counts = Counter(all_words)

top_words_df = pd.DataFrame(
    word_counts.most_common(20),
    columns=["Word", "Count"]
)

top_words_df
```

```python
extra_stop_words = {
    "airline", "again", "about", "review", "then", "fly", "what",
    "verified", "told", "plane", "passengers", "new"
}

stop_words = stop_words.union(extra_stop_words)
```

```python
recommended_reviews = clean_df[
    clean_df["Recommendation_Status"] == "Recommended"
]["Clean_Review"]
```

```python
all_words_recommended = []

for review in recommended_reviews:
    review = re.sub(r"[^a-zA-Z\s]", " ", review)
    words = review.lower().split()

    for word in words:
        if word not in stop_words and len(word) > 2:
            all_words_recommended.append(word)

recommended_word_counts = Counter(all_words_recommended)

top_recommended_words = pd.DataFrame(
    recommended_word_counts.most_common(20),
    columns=["Word", "Count"]
)

top_recommended_words
```

```python
not_recommended_reviews = clean_df[
    clean_df["Recommendation_Status"] == "Not Recommended"
]["Clean_Review"]
```

```python
all_words_not_recommended = []

for review in not_recommended_reviews:
    review = re.sub(r"[^a-zA-Z\s]", " ", review)
    words = review.lower().split()

    for word in words:
        if word not in stop_words and len(word) > 2:
            all_words_not_recommended.append(word)

not_recommended_word_counts = Counter(all_words_not_recommended)

top_not_recommended_words = pd.DataFrame(
    not_recommended_word_counts.most_common(20),
    columns=["Word", "Count"]
)

top_not_recommended_words
```
